```python
import pandas as pd
df = pd.read_csv("..\\data\\processed\\churn_eda.csv")
```

```python
left = df[df['Churn']=='Yes']
stayed = df[df['Churn']=='No']
left_percentage = (len(left)/len(df))*100
stayed_percentage = (len(stayed)/len(df))*100
print("Left: ",len(left),"\nLeft Percentage: ",left_percentage)
print("Stayed: ",len(stayed),"\nStayed Percentage: ",stayed_percentage)
#
```

```python
print("\n\nAverage Tenure For All Customers: ",df['tenure'].mean())
print("Average Tenure For Left Customers: ",left['tenure'].mean())
print("Average Tenure For Stayed Customers: ",stayed['tenure'].mean())
#Customers who churned tend to have much lower tenure.

print("\n\nAverage Monthly Charges For All Customers: ",df['MonthlyCharges'].mean())
print("Average Monthly Charges For Left Customers: ",left['MonthlyCharges'].mean())
print("Average Monthly Charges For Stayed Customers: ",stayed['MonthlyCharges'].mean())
#Customers who churned tend to have higher monthly charges.

print("\n\nAverage Total Charges For All Customers: ",df['TotalCharges'].mean())
print("Average Total Charges For Left Customers: ",left['TotalCharges'].mean())
print("Average Total Charges For Stayed Customers: ",stayed['TotalCharges'].mean())
#Customers who churned tend to have lower total charges, likely because they have lower tenure.
```

```python
male=df[(df['gender']=='Male')]
male_churned=male[male['Churn']=='Yes']
female=df[df['gender']=='Female']
female_churned=female[female['Churn']=='Yes']   
male_percentage=(len(male_churned)/len(male))*100
female_percentage=(len(female_churned)/len(female))*100
print("\n\nMale Churned: ",len(male_churned),"\nMale Churned Percentage: ",male_percentage)
print("Female Churned: ",len(female_churned),"\nFemale Churned Percentage: ",female_percentage)
#Female customers have a slightly higher churn rate than male customers. 
```

```python
senior_yes=df[(df['SeniorCitizen']==1)]
senior_yes_churned=senior_yes[senior_yes['Churn']=='Yes']
senior_no=df[df['SeniorCitizen']==0]
senior_no_churned=senior_no[senior_no['Churn']=='Yes']   
senior_yes_percentage=(len(senior_yes_churned)/len(senior_yes))*100
senior_no_percentage=(len(senior_no_churned)/len(senior_no))*100
print("\n\nSenior Citizens Churned: ",len(senior_yes_churned),"\nSenior Citizens Churned Percentage: ",senior_yes_percentage)
print("Non-Senior Citizens Churned: ",len(senior_no_churned),"\nNon-Senior Citizens Churned Percentage: ",senior_no_percentage)
#Senior citizens have a significantly higher churn rate than non-senior citizens
```

```python
contract_monthly=df[df['Contract']=='Month-to-month']
contract_monthly_churned=contract_monthly[contract_monthly['Churn']=='Yes']
contract_one_year=df[df['Contract']=='One year']
contract_one_year_churned=contract_one_year[contract_one_year['Churn']=='Yes']
contract_two_year=df[df['Contract']=='Two year']
contract_two_year_churned=contract_two_year[contract_two_year['Churn']=='Yes']
contract_monthly_percentage=(len(contract_monthly_churned)/len(contract_monthly))*100
contract_one_year_percentage=(len(contract_one_year_churned)/len(contract_one_year))*100
contract_two_year_percentage=(len(contract_two_year_churned)/len(contract_two_year))*100
print("\n\nMonth-to-month Contract Churned: ",len(contract_monthly_churned),"\nMonth-to-month Contract Churned Percentage: ",contract_monthly_percentage)
print("One year Contract Churned: ",len(contract_one_year_churned),"\nOne year Contract Churned Percentage: ",contract_one_year_percentage)
print("Two year Contract Churned: ",len(contract_two_year_churned),"\nTwo year Contract Churned Percentage: ",contract_two_year_percentage)
#Customers with month-to-month contracts have a much higher churn rate than customers with one year or two year contracts.
```

```python
internet_service_no=df[df['InternetService']=='No']
internet_service_no_churned=internet_service_no[internet_service_no['Churn']=='Yes']
internet_service_dsl=df[df['InternetService']=='DSL']       
internet_service_dsl_churned=internet_service_dsl[internet_service_dsl['Churn']=='Yes']
internet_service_fiber_optic=df[df['InternetService']=='Fiber optic']
internet_service_fiber_optic_churned=internet_service_fiber_optic[internet_service_fiber_optic['Churn']=='Yes']
internet_service_no_percentage=(len(internet_service_no_churned)/len(internet_service_no))*100
internet_service_dsl_percentage=(len(internet_service_dsl_churned)/len(internet_service_dsl))*100
internet_service_fiber_optic_percentage=(len(internet_service_fiber_optic_churned)/len(internet_service_fiber_optic))*100
print("\n\nNo Internet Service Churned: ",len(internet_service_no_churned),"\nNo Internet Service Churned Percentage: ",internet_service_no_percentage)
print("DSL Internet Service Churned: ",len(internet_service_dsl_churned),"\nDSL Internet Service Churned Percentage: ",internet_service_dsl_percentage)
print("Fiber Optic Internet Service Churned: ",len(internet_service_fiber_optic_churned),"\nFiber Optic Internet Service Churned Percentage: ",internet_service_fiber_optic_percentage)
#Customers with fiber optic internet service have a much higher churn rate than customers with DSL or no
```

```python
tech_support_no=df[df['TechSupport']=='No']
tech_support_no_churned=tech_support_no[tech_support_no['Churn']=='Yes']
tech_support_yes=df[df['TechSupport']=='Yes']
tech_support_yes_churned=tech_support_yes[tech_support_yes['Churn']=='Yes']
tech_support_no_percentage=(len(tech_support_no_churned)/len(tech_support_no))*100
tech_support_yes_percentage=(len(tech_support_yes_churned)/len(tech_support_yes))*100
print("\n\nNo Tech Support Churned: ",len(tech_support_no_churned),"\nNo Tech Support Churned Percentage: ",tech_support_no_percentage)
print("Yes Tech Support Churned: ",len(tech_support_yes_churned),"\nYes Tech Support Churned Percentage: ",tech_support_yes_percentage)
#Customers without tech support have a much higher churn rate than customers with tech support.
```

```python
payment_method_electronic_check=df[df['PaymentMethod']=='Electronic check']
payment_method_electronic_check_churned=payment_method_electronic_check[payment_method_electronic_check['Churn']=='Yes']
payment_method_mailed_check=df[df['PaymentMethod']=='Mailed check']
payment_method_mailed_check_churned=payment_method_mailed_check[payment_method_mailed_check['Churn']=='Yes']        
payment_method_bank_transfer=df[df['PaymentMethod']=='Bank transfer (automatic)']
payment_method_bank_transfer_churned=payment_method_bank_transfer[payment_method_bank_transfer['Churn']=='Yes']
payment_method_credit_card=df[df['PaymentMethod']=='Credit card (automatic)']
payment_method_credit_card_churned=payment_method_credit_card[payment_method_credit_card['Churn']=='Yes']   
payment_method_electronic_check_percentage=(len(payment_method_electronic_check_churned)/len(payment_method_electronic_check))*100
payment_method_mailed_check_percentage=(len(payment_method_mailed_check_churned)/len(payment_method_mailed_check))*100
payment_method_bank_transfer_percentage=(len(payment_method_bank_transfer_churned)/len(payment_method_bank_transfer))*100
payment_method_credit_card_percentage=(len(payment_method_credit_card_churned)/len(payment_method_credit_card))*100
print("\n\nElectronic Check Churned: ",len(payment_method_electronic_check_churned),"\nElectronic Check Churned Percentage: ",payment_method_electronic_check_percentage)
print("Mailed Check Churned: ",len(payment_method_mailed_check_churned),"\nMailed Check Churned Percentage: ",payment_method_mailed_check_percentage)
print("Bank Transfer Churned: ",len(payment_method_bank_transfer_churned),"\nBank Transfer Churned Percentage: ",payment_method_bank_transfer_percentage)
print("Credit Card Churned: ",len(payment_method_credit_card_churned),"\nCredit Card Churned Percentage: ",payment_method_credit_card_percentage)
#Customers who pay with electronic checks have a much higher churn rate than customers who pay with mailed checks, bank transfers, or credit cards.
```

```python
paperless_billing_no=df[df['PaperlessBilling']=='No']
paperless_billing_no_churned=paperless_billing_no[paperless_billing_no['Churn']=='Yes']
paperless_billing_yes=df[df['PaperlessBilling']=='Yes']
paperless_billing_yes_churned=paperless_billing_yes[paperless_billing_yes['Churn']=='Yes']
paperless_billing_no_percentage=(len(paperless_billing_no_churned)/len(paperless_billing_no))*100
paperless_billing_yes_percentage=(len(paperless_billing_yes_churned)/len(paperless_billing_yes))*100
print("\n\nNo Paperless Billing Churned: ",len(paperless_billing_no_churned),"\nNo Paperless Billing Churned Percentage: ",paperless_billing_no_percentage)
print("Yes Paperless Billing Churned: ",len(paperless_billing_yes_churned),"\nYes Paperless Billing Churned Percentage: ",paperless_billing_yes_percentage)
#Customers with paperless billing have a much higher churn rate than customers without paperless billing.
```
